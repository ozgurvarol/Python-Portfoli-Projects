```python
import numpy as np
import pandas as  pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
###  Battle_Week09.csv  -  Displays the battle winners and losers
df1 = pd.read_csv("Battle_Week09.csv")
df1
```

```python
### 1)  For the winners, what is the top 5 most and least common 'Type 1'?

Battle = pd.merge(df1,df2,on='#',how='inner')
Battle
```

```python
Winner = Battle[Battle['Win']==1]
Winner

```

```python
dfType1Winner = Winner.groupby(['Type 1'])
print(dfType1Winner['Win'].count())
```

```python
Least5WinnerType1 = Winner['Type 1'].value_counts()[-5:] 
Least5WinnerType1

```
