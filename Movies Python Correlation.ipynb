```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\ozgur\Desktop\movies.csv')
df.head()
```

```python
## I am checking if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
# I had to fill all NA values for 0 to able to change data type.

df = df.fillna(0)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df  
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
df.drop_duplicates()

#checking if there is any duplicate values.
```

```python
## Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
## Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data=df, scatter_kws = {"color": "red"}, line_kws={"color": "blue"})
```

```python
## Correlation #My hypothesis was there is a high correlation between gross and budget. and this is the proof base on pearson correlation

df.corr()
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Feature')

plt.show()
```

```python


df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
## Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# I can now take a look at the ones that have a high correlation (> 0.5)


strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]


print(strong_pairs)
```
